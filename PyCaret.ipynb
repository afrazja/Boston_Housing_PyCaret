Regression

```python
import pandas as pd

df = pd.read_csv('sample_data/california_housing_train.csv')

df
```

```python
from pycaret.regression import *

s = setup(df, target='median_house_value')
```

```python
best = compare_models()
```

```python
print(best)
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
test_df = pd.read_csv('sample_data/california_housing_train.csv')

predict_model(best, test_df)
```

Classification

```python
titanic_train = pd.read_csv('train.csv')
titanic_test = pd.read_csv('test.csv')

titanic_train
```

```python
titanic_train.drop('Survived', axis=1, inplace=True)
```

```python
titanic_train['Pclass'].value_counts()
```

```python
from pycaret.classification import *

s = setup(titanic_train, target = 'Pclass')
```

```python
best = compare_models()
```

```python
print(best)
```

```python
finalize_model(best)
```

```python
evaluate_model(best)
```

```python
predict_model(best)
```

```python
predict_model(best, titanic_test)
```
